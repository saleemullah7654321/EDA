```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv('dataset/titanic.csv')
tips = sns.load_dataset(name='tips')
flights = sns.load_dataset(name='flights')
iris = sns.load_dataset(name='iris')
```

```python
tips.sample(2)
```

```python
flights.sample(2)
```

```python
iris.sample(2)
```

### 1. ScatterPlot (NUmerical - Numerical) relationship

```python
# as shows by increasing bill tip also increases
sns.scatterplot(tips['total_bill'],tips['tip'])
```

```python
# understand which gender give more tip analyze multi col gender, tip ans total_bill
sns.scatterplot(tips['total_bill'],tips['tip'],hue=tips['sex'])
```

```python
# i want to analyze 1 more col to find smoker as well
sns.scatterplot(tips['total_bill'],tips['tip'],
                hue=tips['sex'],style=tips['smoker'])
```

```python
# also want to add another col size
sns.scatterplot(tips['total_bill'],tips['tip'],
                hue=tips['sex'],style=tips['smoker'],
                size=tips['size'])
```

### 4. Distplot (NUmerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

### 5. HeatMap (Categorical - Categorical)

```python
data = pd.crosstab(titanic['Pclass'],titanic['Survived'])
data
```

```python
# light colors indicated high count like bottom left shows light color for not_survived
# on the other hand top right also light indicating many people survived in 1 class category
sns.heatmap(data=data)
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
for c in ['Sex','Embarked']:
    print(titanic.groupby(c).mean()['Survived']*100,'\n\n')
```
